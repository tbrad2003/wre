# Centroids

The location of the centroid of an area, bounded by the axes and the function $y=f(x)$, can be found by integration. 

$$ x_c = \frac{\int xdA}{A} $$
$$ y_c = \frac{\int ydA}{A} $$
$$ A = \int f(x)dx $$
$$ dA = f(x)dx = g(y)dy $$
